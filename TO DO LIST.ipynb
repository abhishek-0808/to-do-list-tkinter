```python
import tkinter                #tool from graphical user interface (GUI)
from tkinter import *         #import all the modules from tkinter library
import os 
from PIL import Image, ImageTk

root = Tk()                   #creating the main window for GUI
root.title("MY TO-DO LIST")   #title for the application

'''
  root.geometry() : 
  root.geometry("widthxheight+XPOS+YPOS")
  width: The width of the window in pixels.
  height: The height of the window in pixels.
  XPOS: The X-coordinate position of the window on the screen.
  YPOS: The Y-coordinate position of the window on the screen.
  
'''

root.geometry("400x650+400+100")
root.resizable(False, False)   #making the main window resizable - this function is used to control if we want to resize the main window

task_list=[]

def addTask():
    task=task_entry.get()
    task_entry.delete(0, END)
    
    if task:
        with open("tasklist.txt", "a") as taskfile:
            taskfile.write(f"\n{task}")
        task_list.append(task)
        listbox.insert(END, task)
        
def deleteTask():

    global task_list
    
    task=str(listbox.get(ANCHOR))
    if task in task_list:
        task_list.remove(task)
        with open("tasklist.txt","w") as taskfie:
            for task in task_list:
                taskfie.write(task + "\n")
        listbox.delete(ANCHOR)
    
def Opentaskfile(): 
    try:
        global task_list
        with open("tasklist.txt","r") as taskfile:
            tasks=taskfile.readlines()

        for task in tasks:
            if task!="\n":
                task_list.append(task)
                listbox.insert(END,task)
                
    except:
        file = open("tasklist.txt", "w")
        file.close()


#Icon for the TO DO List
Image_icon=PhotoImage(file=r"C:\Users\user\Desktop\Images\task.png")     #a PhotoImage object is being created where we are storing the image file
root.iconphoto(False, Image_icon)      #iconphoto method is used to set the icon of the tkinter window

#Top bar
TopImage=PhotoImage(file=r"C:\Users\user\Desktop\Images\topbar.png")
Label(root, image = TopImage).pack()

dockimage = PhotoImage(file = r"C:\Users\user\Desktop\Images\dock.png")
Label(root, image=dockimage, bg="#87CEEB").place(x=30, y =25)

'''This line creates a Tkinter Label widget, associates the dockimage with it, 
   sets the background color (bg) of the label to "#87CEEB" (a hexadecimal color code), 
   and then uses the place geometry manager to specify the position of the label within the Tkinter window.'''


noteimage= PhotoImage(file = r"C:\Users\user\Desktop\Images\task.png")
Label(root, image=noteimage, bg="#87CEEB").place(x=340, y =25)

heading = Label(root , text= "ALL TASKS", font="Helvetica 16 bold" , fg = "#3498db" , bg="#e74c3c").place(x=130, y = 20)

#main program

frame= Frame(root, width=400, height = 50, bg="white" )
frame.place(x=0, y=180)

task= StringVar()
task_entry = Entry(frame, width = 18, font="Helvetica 16", bd=0)
task_entry.place(x=10, y=7)
task_entry.focus()

button=Button(frame, text="ADD ITEM" , font="Helvetica 12 bold" , width=10, bg="#5a95ff", fg="#fff", bd=0, command = addTask)
button.place(x=300, y=10)

#creating the listbox
frame1= Frame(root, bd=3, width = 700 , height = 300, bg="#32405b")
frame1.pack(pady=(160,0))
listbox=Listbox(frame1, font=("Helvetica ",12), width =40 , height = 16 , bg="#32405b", fg="white", cursor="hand2", selectbackground="#5a95ff")
listbox.pack(side = LEFT , fill = BOTH , padx=2)

scrollbar= Scrollbar(frame1)
scrollbar.pack(side = RIGHT , fill = BOTH)

listbox.config(yscrollcommand = scrollbar.set)
scrollbar.config(command=listbox.yview)

Opentaskfile()



#delete button
Delete_icon=PhotoImage(file = r"C:\Users\user\Desktop\Images\delete.png")
Button(root, image= Delete_icon, bd=0,command= deleteTask).pack(side=BOTTOM , pady=2)


root.mainloop()
```
